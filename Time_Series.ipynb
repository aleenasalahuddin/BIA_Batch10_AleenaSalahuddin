```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/BIA_Batch10_Datasets/yahoo_stock.csv')
```

```python
date = df['Date']
```

```python
closing = df['Close']
```

```python
plt.plot(date, closing)
plt.title('Closing price by date')
step = 365
plt.xticks(date[::step], rotation=30)
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.show
```

```python
df['moving_avg'] = df['Close'].rolling(window=100).mean()

plt.plot(date, closing, label='Closing Price')
plt.plot(date, df['moving_avg'], label='100-day Moving Avg', color='orange')

plt.title('Closing Price with Moving Average')
step = 365
plt.xticks(date[::step], rotation=30, ha='right')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['daily_change'] = df['Close'].diff()

# Plot daily change
plt.plot(date, df['daily_change'], label='Daily Change', color='red')

plt.title('Daily Price Change')
step = 100
plt.xticks(date[::step], rotation=30, ha='right')
plt.xlabel('Date')
plt.ylabel('Change in Price')
plt.legend()
plt.show()
```

# **Analysis**
Based on the provided plots of Yahoo stock data, the closing price generally shows an upward trend over the observed period, indicating overall growth. The 100-day moving average smooths out the daily fluctuations and confirms this upward trend, also highlighting periods of more rapid increase and occasional dips. The daily price change plot reveals the day-to-day volatility of the stock, with most changes being relatively small but with occasional significant positive or negative spikes. These spikes could correspond to specific market events or news affecting the stock. In conclusion, while the stock demonstrates an overall positive trajectory, it is subject to notable short-term price swings.
