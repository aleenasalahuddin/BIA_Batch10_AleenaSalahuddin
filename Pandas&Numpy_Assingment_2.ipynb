3.  **Create a function that adds an 'age_group' column to df_customers, categorizing ages as 'Young' (<30), 'Adult' (30-40), or 'Senior' (>40), and returns the updated DataFrame.**

```python
def age_group(df):
    conditions = [
        df['age'] < 30,
        (df['age'] >= 30) & (df['age'] <= 40),
        df['age'] > 40
    ]
    choices = ['Young', 'Adult', 'Senior']
    df['age_group'] = np.select(conditions, choices, default='Unknown')
    return df
```

4. **Create a function that normalizes the 'salary' column in df_employees to a 0-1 scale and returns the DataFrame with a new 'normalized_salary' column.**

```python
def normalize_salary(df):
    maximum = df['salary'].max()
    minimum = df['salary'].min()

    df['normalized'] = (df['salary'] - minimum) / (maximum - minimum)
    return df
```

```python
normalized_salary = normalize_salary(df_employees)
normalized_salary
```

```python
# Function: add_salary_rank
# Create a function that adds a 'salary_rank' column to df_employees, ranking employees
# by salary (highest salary = rank 1), and returns the updated DataFrame.
# DataFrame: df_employees
def add_salary_rank(df):
    df['salary_rank'] = df['salary'].rank(ascending=False, method='dense').astype(int)
    return df
```

```python
#56
# All the values in df_customers are clean so I'm appending some rough values in age cell
customers = {
    "customer_id" : [107, 108, 109, 110],
    "name" : ['Grey', 'Harry', 'Isabella', 'Jhonas'],
    "age": [' ', np.nan, np.nan, ''],
    "city": ['Boston', 'Dallas', 'Seattle', 'Atlanta']
}
df_more_customers = pd.DataFrame(customers)

df_customers = pd.concat([df_customers, df_more_customers])

# Function: fill_missing_ages
# Create a function that fills missing values in the 'age' column of df_customers with the
# mean age and returns the updated DataFrame.
# DataFrame: df_customers
def fill_missing_ages(df_customers):
    df_customers['age'] = df_customers['age'].replace(r'^\s*$', np.nan, regex=True)
    df_customers['age'] = df_customers['age'].astype(float)
    df_customers['age'] = df_customers['age'].fillna(df_customers['age'].mean()).round().astype(int)
    return df_customers

fill_missing_ages(df_customers)
```
